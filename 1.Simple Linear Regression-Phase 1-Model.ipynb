```python
#import pandas library to work with datasheets.
import pandas as pd
```

```python
#using padas read_csv method we are reading the data from " Salary_Data.csv " file.
dataset = pd.read_csv("Salary_Data.csv")
```

```python
#here we are displaying the data which assigned to dataset variable.
dataset
```

```python
#here we are assigning years of experience column values to independent variable.
independent = dataset["YearsExperience"]
```

```python
#Displaying the independent variable data
independent
```

```python
#to the depenedent variable, assigning dataset Salary column values.
dependent = dataset[["Salary"]]
```

```python
#displaying dependent variable data
dependent
```

```python
#from sklearn library model selection, we are importing train_test_split functionality.
from sklearn.model_selection import train_test_split
```

```python
#to the train_test_split method we are giving inputs/learning data in the sequence like training, test data, test data percentage as 30% and random_state as zero.
x_train, x_test, y_train, y_test = train_test_split(independent, dependent, test_size = 0.30, random_state = 0)
```

```python
#from sklearn linear_model module we are importing LinearRegression here.
#to the regressor varibale we are assigning LinearRegression() method values
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```
