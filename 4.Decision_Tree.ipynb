```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("50_Startups.csv")
```

```python
dataset
```

```python
dataset.columns
```

```python
#converting text data to numeric using dummies method
dataset = pd.get_dummies(dataset, drop_first = True)
```

```python
#displaying data post the conversion
dataset
```

```python
#assigning independent, dependent values
independent=dataset[['R&D Spend','Administration', 'Marketing Spend','State_Florida', 'State_New York']]
dependent=dataset[['Profit']]
```

```python
#displaying independent assigned variable values.
independent
```

```python
#displaying dependent variable values
dependent
```

```python
#split into training set and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(independent, dependent, test_size = 0.30, random_state = 0)
```

```python
X_train
```

```python
#from sklearn linear_model module we are importing LinearRegression here.
#to the regressor varibale we are assigning LinearRegression() method values
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(criterion = 'absolute_error', splitter = 'best')
regressor = regressor.fit(X_train, y_train)
```

```python
regressor
```

```python
from sklearn import tree
tree.plot_tree(regressor)
plt.show()
```

```python
y_pred = regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score = r2_score(y_test, y_pred)
```

```python
r_score
```

```python
#importing pickle library for deployment phase
import pickle
```

```python
#here we are assigning the saving model file name with extension to the filename variable
filename = "finalized-model_DecisionTree_AbsoluteError_Best.sav"
#using pickle dump method we are writing the file on disk
pickle.dump(regressor, open(filename, 'wb'))
```

```python
#loading the saved model to loaded_variable.
loaded_model = pickle.load(open("finalized-model_DecisionTree_AbsoluteError_Best.sav", 'rb'))
```

```python
#here we are pedicting the value like we want and assigning to result variable.
result = loaded_model.predict([[1234, 345, 4565, 1, 0]])
```
